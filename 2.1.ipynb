Подключаем нужные нам библиотеки numpy, matplotlib и scipy.stats.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sps
```

Определяем размер нашей выборки N и параметр $\theta$ (в коде обозначено как t). Генерируем выборку из равномерного распределения на отрезке [0;$\theta$] размера N = 10000. Далее задаём массивы, заполненные нулями, в которых мы будем хранить наши оценки. 

```python
t = 25
n = 10000
sample = sps.uniform.rvs(size=n, loc=0, scale=t)
arr = np.arange(0, n, 1)
first_ev = np.zeros(n)
second_ev = np.zeros(n)
third_ev = np.zeros(n)
fourth_ev = np.zeros(n)
fifth_ev = np.zeros(n)
```

Заполняем наши массивы first_ev, second_ev, third_ev, fourth_ev и fifth_ev как модуль разности оценки(среза выборки) и истинного значения $\theta$ в зависимости от n. 

```python
for i in range(1, n):
    first_ev[i] = abs(2 * np.mean(sample[:i]) - t)
    second_ev[i] = abs(np.mean(sample[:i]) + np.max(sample[:i]) / 2 - t)
    third_ev[i] = abs((i + 1) * np.min(sample[:i]) - t)
    fourth_ev[i] = abs(np.min(sample[:i]) + np.max(sample[:i]) - t)
    fifth_ev[i] = abs(np.max(sample[:i]) * (i + 1) / i - t)
```

Строим на одном графике разными цветами для всех оценок функции модуля разности оценки и истинного значения $\theta$ в зависимости от n. Для избавления от больших значений разности в начале ограничиваем масштаб графика, для наглядности соединяем точки линиями.

```python
plt.figure(figsize=(20, 10))
plt.subplot(211)
plt.ylim(0, 2)
plt.plot(arr, first_ev, label='first_ev')
plt.plot(arr, second_ev, label='second_ev')
plt.plot(arr, third_ev, label='third_ev')
plt.plot(arr, fourth_ev, label='fourth_ev')
plt.plot(arr, fifth_ev, label='fifth_ev')
plt.legend()
plt.show()
```

Чтобы понять, какая оценка получилась лучше, проводим эксперимент для различных значений $\theta$. В дальнейшем исключаем третью оценку, так как она сильно отличается от истинного значения $\theta$. При $\theta$ = 50:
